```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-v0_8')
```

```python
df=pd.read_csv('SuperStore_Sales_Dataset.csv')
df.head()
```

```python
print('Number of Rows:', df.shape[0])
print('Number of Columns:', df.shape[1])
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
print('Drop Duplicate:', df.duplicated().sum())
df=df.drop_duplicates()
print('After Removing Duplicate:', df.duplicated().sum())
```

```python
Total_sales=df['Sales'].sum()
Total_profit=df['Profit'].sum()
revenue=(Total_sales-Total_profit)
print('Total Sales:', Total_sales)
print('Total Profit:', Total_profit)
print('Revenue:', revenue)
```

```python
Total_Qty=df['Quantity'].sum()
Total_return=df['Returns'].sum()
Overall_Sales=(Total_Qty-Total_return)
print('Total_Quantity:', Total_Qty)
print('Total_Retruns:', Total_return)
print('Overall_Sales:', Overall_Sales)
```

```python
Top_category=df['Category'].value_counts()
plt.figure(figsize=(7,3))
sns.barplot(x=Top_category.values, y=Top_category.index, palette='viridis')
plt.title('Top Categories')
plt.xlabel('Sales')
plt.ylabel('Category Name')
plt.show()
```

```python
Segment=df['Segment'].value_counts()
plt.figure(figsize=(7,4))
plt.pie(Segment, labels=Segment.index, autopct='%1.1f%%', colors=['green', 'orange','skyblue'])
plt.title('Top Sengments')
plt.show()
```

```python
Top_city=df['City'].value_counts().head(10)
plt.figure(figsize=(10,4))
sns.barplot(x=Top_city.values, y=Top_city.index, palette='viridis')
plt.title('Top 10 City by Sales')
plt.xlabel('Sales')
plt.ylabel('City Name')
plt.show()
```

```python
Top_state=df['State'].value_counts().head(10)
plt.figure(figsize=(10,4))
sns.barplot(x=Top_state.values, y=Top_state.index, palette='viridis')
plt.title('Top 10 State')
plt.xlabel('Sales')
plt.ylabel('State Name')
plt.show()
```

```python
Payment_mode=df['Payment Mode'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(Payment_mode, labels=Payment_mode.index, autopct='%1.1f%%', colors=['orange', 'green', 'red'], shadow=True)
plt.title('Payment Mode')
plt.show()
```

```python
Ship_mode=df['Ship Mode'].value_counts()
plt.figure(figsize=(10,5))
plt.pie(Ship_mode, labels=Ship_mode.index, autopct='%1.1f%%', wedgeprops=dict(width=0.5), colors=['orange','green','yellow','pink'], shadow=True)
plt.title('Top Ship Mode')
plt.show()
```

```python
Sub_category=df['Sub-Category'].value_counts().head(5)
plt.figure(figsize=(10,5))
sns.barplot(x=Sub_category.values, y=Sub_category.index, palette='viridis')
plt.title('Top 5 Sub_Category', color='red')
plt.xlabel('Sales')
plt.ylabel('Sub_category_name')
plt.show()
```

```python
Top_customer=df['Customer Name'].value_counts().head(5)
plt.figure(figsize=(7,5))
sns.barplot(x=Top_customer.values, y=Top_customer.index, palette='viridis')
plt.title('Top 5 Customers')
plt.xlabel('Count')
plt.ylabel('Customer Name')
plt.show()
```

```python
Top_region=df['Region'].value_counts()
plt.figure(figsize=(7,5))
plt.pie(Top_region, labels=Top_region.index, autopct='%1.1f%%', shadow=True)
plt.title('Region Wise Sales')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='Sales', y='Order Date', data=df, color='green')
plt.title('Date Vs Sales')
plt.xlabel('Sales')
plt.ylabel('Date')
plt.show()
```
